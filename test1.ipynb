```python
import matplotlib.patches as mpatches
from scipy import signal
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np
from tempfile import NamedTemporaryFile

VIDEO_TAG = """<video controls>
 <source src="data:video/x-m4v;base64,{0}" type="video/mp4">
 Your browser does not support the video tag.
</video>"""

def anim_to_html(anim):
    if not hasattr(anim, '_encoded_video'):
        with NamedTemporaryFile(suffix='.mp4') as f:
            anim.save(f.name, fps=20, extra_args=['-vcodec', 'libx264'])
            video = open(f.name, "rb").read()
        anim._encoded_video = video.encode("base64")
    
    return VIDEO_TAG.format(anim._encoded_video)


from IPython.display import HTML

def display_animation(anim):
    plt.close(anim._fig)
    return HTML(anim_to_html(anim))


from matplotlib import animation



############################# startint value #####################################3
G = a.subs(T,3)/b.subs(T,3)

a1, b1 = G.as_numer_denom()
a1 = np.array(Poly(a1).coeffs(),dtype=np.float64)
b1 = np.array(Poly(b1).coeffs(),dtype=np.float64)
s1 = signal.lti(a1, b1)
w, mag, phase = s1.bode(w=[1e-1,1e3])

###############################################3
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.set_xlabel('w')
ax1.set_ylabel('gain')

line1 = Line2D([], [], color='black',label="batata")

ax1.add_line(line1)
ax1.semilogx(w,mag)
ax1.set_ylim([-30,30])
ax1.set_xlim([10e-1,1e3])

###############################################
ax2.set_xlabel('omega')
ax2.set_ylabel('theta')
line2 = Line2D([], [], color='black',label="test1")
ax2.add_line(line2)

ax2.semilogx(w, phase)  # Bode phase plot
ax2.set_ylim([-180,180])

ax2.set_xlim([10e-1,1e3])
legend = ax1.legend(loc='upper center', shadow=True)


# initialization function: plot the background of each frame

# animation function.  This is called sequentially
def animate(i): 
    k = i/10.0 - 10
    G = a.subs(T,k)/b.subs(T,k)
    
    a1, b1 = G.as_numer_denom()
    try:
        a1 = np.array(Poly(a1).coeffs(),dtype=np.float64)
    except:
        a1 = np.array(a1,dtype=np.float64)
    b1 = np.array(Poly(b1).coeffs(),dtype=np.float64)
    s1 = signal.lti(a1, b1)
    w, mag, phase = s1.bode(w=[1e-1,1e3])
    
    line1.set_data(w, mag)
    line2.set_data(w, phase)
    line1.set_label("k={0}".format(k))
    legend = ax1.legend(loc='upper center', shadow=True)
    #return line,

# call the animator.  blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, frames=200, interval=50, blit=True)

# call our new function to display the animation
display_animation(anim)
```
